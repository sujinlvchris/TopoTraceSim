## Mapping

```python
import torch
import os
```

### Default mapping is based on heuristic.

```python
device = torch.device("npu:0")

input = torch.randn(1024, 1024).to(device=device)
weight = torch.randn(1024, 1024).to(device=device)

opt_fn = torch.compile(dynamic=False)(torch.matmul)
npu_outut = opt_fn(input, weight)
```

```python
log_path = ""
!cat $log_path | grep "Total execution cycle"
```

### External Mapping
User can set tile size manually from external file.

```python
torch._dynamo.reset()

os.environ['TOGSIM_CONFIG']="/workspace/PyTorchSim/tutorial/session1/togsim_configs/togsim_config_external_mapping.yml"

input = torch.randn(1024, 1024).to(device=device)
weight = torch.randn(1024, 1024).to(device=device)

opt_fn = torch.compile(dynamic=False)(torch.matmul)
npu_outut = opt_fn(input, weight)
```
