## Execution Mode

```python
import torch
import os
```

### Functional & Timing mode (Default)

```python
device = torch.device("npu:0")

input = torch.randn(512, 512).to(device=device)
weight = torch.randn(512, 512).to(device=device)

opt_fn = torch.compile(dynamic=False)(torch.matmul)
npu_output = opt_fn(input, weight)
```
